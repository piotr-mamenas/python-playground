```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,20,50)
y = x**2
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,15)
plt.ylim(0,250)
plt.title("PLOT")
plt.xlabel("X label")
plt.ylabel("Y label")
```
